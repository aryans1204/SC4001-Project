## Input Examples

```python
import torch
from torchvision import utils
import matplotlib.pyplot as plt
import numpy as np
from data import celebds, utkfaceds, GENDERS


def imshow(inp, title=None):
    inp = inp.numpy().transpose((1, 2, 0))
    mean = np.array([0.485, 0.456, 0.406])
    std = np.array([0.229, 0.224, 0.225])
    inp = std * inp + mean
    inp = np.clip(inp, 0, 1)
    plt.imshow(inp)

    if title is not None:
        plt.title(title)

    plt.pause(0.001)
```

### UTKFace Dataset

```python
utkfaceds.setup()

inputs, labels = next(iter(utkfaceds.test_dataloader()))
out = utils.make_grid(inputs)

imshow(out, title=[GENDERS[int(label)] for label in labels])
```

### CelebA Dataset

```python
celebds.setup()

inputs, labels = next(iter(celebds.test_dataloader()))
out = utils.make_grid(inputs)

imshow(out, title=[GENDERS[int(label)] for label in labels])
```
